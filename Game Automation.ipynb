```python
"""   setting all the rgb values to ZERO gives black color, here we do the same using np.zeros_like()   """ 
my_img = np.zeros_like(img_crop)


"""    visualise black canvas   """

plt.imshow(my_img)
```

```python
rcord1=set()


for j in range(y_crop):
    for k in range(x_crop):
        if ((img_crop[j,k,0]==255) and (img_crop[j,k,1]==0) and (img_crop[j,k,2]==0)):
            my_img[j,k,0] = 0
            my_img[j,k,1] = 255
            my_img[j,k,2] = 0
            rcord1.add(k)  # record the x-coordinates of your red-color cells in a set
```

```python
rcord2=set()


for j in range(y_crop):
    for k in range(x_crop):
        if ((img_crop[j,k,0]==247)and (img_crop[j,k,1]==27)and (img_crop[j,k,2]==27)):
            my_img[j,k,0] = 0
            my_img[j,k,1] = 255
            my_img[j,k,2] = 0
            rcord2.add(k)  # record the x-coordinates of your red-color cells in a set
```

```python
dist = np.ceil(abs(src-target))
rate = 1.05
time = str(int(dist/rate))

print(dist,time)
```

```python
rate = 1.05

while True:

    os.system('adb shell screencap -p /sdcard/screencap.png')
    os.system('adb pull /sdcard/screencap.png')

    # read the image using openCV
    img = cv2.imread('screencap.png')
    screen_height, screen_width, _ = img.shape

    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    img_crop = img[ 1450:1600, :]
    y_crop, x_crop, z_crop = np.shape(img_crop)
    
    """   setting all the rgb values to ZERO gives black color, here we do the same using np.zeros_like()   """ 
    my_img = np.zeros_like(img_crop)

    rcord1=set()
    
    for j in range(y_crop):
        for k in range(x_crop):
            if ((img_crop[j,k,0]==255) and (img_crop[j,k,1]==0) and (img_crop[j,k,2]==0)):
                my_img[j,k,0] = 0
                my_img[j,k,1] = 255
                my_img[j,k,2] = 0
                rcord1.add(k)  # record the x-coordinates of your red-color cells in a set
    
    rcord2=set()
    
    for j in range(y_crop):
        for k in range(x_crop):
            if ((img_crop[j,k,0]==247)and (img_crop[j,k,1]==27)and (img_crop[j,k,2]==27)):
                my_img[j,k,0] = 0
                my_img[j,k,1] = 255
                my_img[j,k,2] = 0
                rcord2.add(k)  # record the x-coordinates of your red-color cells in a set
    
    # sort the elements in the set in ascending order and make a list

    src = statistics.mean(rcord1)
    target = statistics.mean(rcord2)
    
    dist = np.ceil(abs(src-target))
    time = str(int(dist/rate))
    
    print(dist)
    
    """   just fix some x, y coordinate to swipe on the screen, with time already calculated before  """ 

    os.system('adb shell input touchscreen swipe 530 530 830 830 '+ time)
    pause.seconds(3)
```
